```python
# read the data from word_freq_5_sp.txt 
# production: 80953
# of: 13236223
# massive: 77379
# photon: 77053
# pairs: 55240
# at: 1074162

import numpy as np
import pandas as pd


def read_data(filename):
    data = pd.read_csv(filename, sep=': ', header=None)
    data.columns = ['word', 'freq']
    return data

df = read_data('word_freq_5_sp_EQU.txt')
df
```

```python
# df['word] all to lower case
df['word'] = df['word'].str.lower()
# add the freq of the same word
df = df.groupby(['word']).sum()
# sort the freq
df = df.sort_values(by=['freq'], ascending=False)
# filt all the freq < 20
df = df[df['freq'] >= 50]


df
```

```python
df.to_csv('word_freq_50_sp.csv', sep=',', header=True, index=True)
wordlist = df.index.tolist()
# CITE , FIG , REF, SEC, POINT add to token list
wordlist.append('CITE')
wordlist.append('FIG')
wordlist.append('REF')
wordlist.append('SEC')
wordlist.append('POINT')
# save the wordlist
with open('wordlist.txt', 'w') as f:
    for item in wordlist:
        f.write("%s\n" % item)





```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer(vocab_file='wordlist.txt')

token_list = []

tokens = tokenizer.tokenize("Coronavirus disease 2019 covid-19 is a contagious disease caused by the virus SARS-CoV-2.")
token_list.extend(tokens)

# 打印 token list
print(token_list)
# save the tokenizer
tokenizer.save_pretrained('/home/jjdong/wholewordbert')
```
